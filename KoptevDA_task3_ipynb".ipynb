# Лабораторная работа №2, задание 3
Графическое построение линий уровня, сравнение с аналитическими расчетами

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(X, C):
    return (C[0]*X[0] + C[1]*X[1]) + 1/3 * pow((pow(X[0], 2) + pow(X[1], 2)), 3/2)
```

```python
# Задаем значения
X = [0,0]
X[0], X[1] = np.meshgrid(np.linspace(-10, 8, 10000), np.linspace(-10, 3, 10000))
C = (5,2)
```

```python
F = f(X, C)
# минимальное значение функции
F.min()
```

```python
# найдем x и y, при которых значение функции будет минимальным
np.where(F == F.min())
```

```python
X[0][7029, 4358], X[1][7029, 4358]
```

```python
# График функции
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X=X[0], Y=X[1], Z=F)
plt.show()
```

```python
# Линии уровня
plt.contourf(X[0], X[1], F, levels=30)
# Добавляем точку минимума
plt.plot(X[0][7029, 4358], X[1][7029, 4358], marker='o')
plt.show()
```
